## 1. 정미니 날짜 데이터 보기

```python
import pandas as pd

# 엑셀 파일 읽기
data = pd.read_excel("제주지방_날짜별평균강수량.xlsx")

# 데이터 확인
print(data)
```

```python
# NAN을 0으로 채우기
data.fillna(0, inplace=True)

# 데이터 확인
print(data)
```

## 2. 내꺼 경기도9개합침 데이터 날짜범위 해당 장마철로 조절

```python
import pandas as pd

# 엑셀 파일 읽기
combined_9 = pd.read_excel("제주도9개합침.xlsx")

# 데이터 확인
print(combined_9)
```

```python
import pandas as pd

# 날짜 범위 설정
start_date = '2022-06-19'
end_date = '2022-07-20'

# 날짜 범위에 해당하는 데이터 필터링
filtered_data = combined_9[(combined_9['날짜'] >= start_date) & (combined_9['날짜'] <= end_date)]
filtered_data = filtered_data.reset_index(drop=True)

# 필터링된 데이터 출력
filtered_data
```

## 3. 정민 data 와 연아 filtered_data 합치기
- 날짜 2개 들어가니까 알맞게 잘 할당되었는지 확인하기

```python
# 두 데이터프레임을 옆으로 합치기
combined_data = pd.concat([data, filtered_data], axis=1)

# 데이터 확인
combined_data
```

```python
combined_data = combined_data.drop(columns='날짜')
combined_data
```

## 엑셀 파일로 저장

```python
# 데이터프레임을 엑셀 파일로 저장
output_file = '날씨랑제주도.xlsx'
combined_data.to_excel(output_file, index=False)

print(f'Data saved to {output_file}')
```
